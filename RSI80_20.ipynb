```python
# def checker(stock_name,start_date,end_date):
df1 = yf.download('RELIANCE.NS', 
                  start='2021-01-01', 
                  end='2021-09-30',
                  interval='1D')
df1.reset_index(inplace=True)
df1=df1.dropna().reset_index(drop=True)
```

```python
df1['RSI'] = talib.RSI(df1['Close'], timeperiod=14) #RSI
```

```python
df1['Range']= ['above' if df1['RSI'][i] > 68 else 'below' if df1['RSI'][i] < 20.2 else 'avoid' for i in range(0,len(df1))]
```

```python
df2=df1[df1['Range'].str.contains('above|below')].reset_index(drop=True)
```
