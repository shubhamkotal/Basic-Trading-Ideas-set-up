```python
def checker(stock_name):
  data = yf.download(stock_name, 
                    start='2021-01-01', 
                    end='2021-11-01')
  # data = yf.download('Reliance.NS', 
  #                   start='2021-01-01', 
  #                   end='2021-11-01')
  data=data.reset_index()
  data['Color'] = ['Red' if data['Open'][i] > data['Close'][i] else 'Green' if data['Open'][i] < data['Close'][i] else 'Neutral' for i in range(0,len(data))]
  value=[]
  for i in range(0,len(data)):
    if data['Color'][i]== 'Red':
      # tail= (data['High'][i] - data['Open'][i]) + (data['Close'][i] - data['Low'][i])
      body= data['Open'][i] - data['Close'][i]
      total= data['High'][i] - data['Low'][i]
      if (body*100)/total > 75:
        candle = 'Sided'
      else:
        candle = 'Sideways'  
      value.append(candle)
    elif data['Color'][i]== 'Green':
      # tail= (data['High'][i] - data['Close'][i]) + (data['Open'][i] - data['Low'][i])
      body= data['Close'][i] - data['Open'][i]
      total= data['High'][i] - data['Low'][i]
      if (body*100)/total > 75:
        candle = 'Sided'
      else:
        candle = 'Sideways'  
      value.append(candle)
    else:
      value.append('Sideways')
  data['Side']=value
  return (len(data[data['Side']=='Sided']))
```

```python
# create df of nested info per stock
selected_stocks={}
for i in tq.tqdm(nse['Symbol']):
    selected_stocks[i]=checker(i)
final_selection=pd.DataFrame.from_dict(selected_stocks,orient='index',columns=['Counts'])
final_selection.reset_index(inplace=True)
final_selection=final_selection.rename(columns={'index':'Ticker'})
final_selection=final_selection.reset_index().drop('index',axis=1)
```
